```python
import os

def search_string_in_files(root_dir, target_string):
    # List to store paths of files containing the target string
    matching_files = []
    
    # Walk through all files and folders within root_dir
    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            try:
                # Open and read each file to search for the target string
                with open(file_path, 'r', encoding='utf-8') as file:
                    if target_string in file.read():
                        matching_files.append(file_path)
            except (UnicodeDecodeError, PermissionError) as e:
                # Skip files that can't be read as text or don't have permissions
                print(f"Skipping file {file_path}: {e}")
    
    # Print results
    if matching_files:
        print("Files containing the target string:")
        for file_path in matching_files:
            print(file_path)
    else:
        print("No files containing the target string were found.")

# Run the function in the current directory
search_string_in_files(".", "url")


```
